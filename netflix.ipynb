```python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")
import re
```

```python
# ジャンルを調べる
# コンマ区切りの文字列を配列にする

genres = []

for i in df['listed_in']:
    genre = re.split(r', \s*', i)
    genres.append(genre)
```

```python
genres
```

```python
# 2次元配列を1次元配列にする
genres_flat = []
for sublist in genres:
    for item in sublist:
        genres_flat.append(item)
```

```python
generes_df = pd.DataFrame(genres_flat)
```

```python
generes_df
```

```python
# 重複削除してAtoZにsort
sorted(set(genres_flat))
```

```python
generes_df[0].value_counts()
```

```python
generes_df[0].value_counts().plot(kind="bar", figsize=(20, 10))
```

```python
df
```

```python
# 国別のジャンルを返す関数
def genres_by_country(c):
    genres = []
    genres_flat = []
    for i in df[df.country.str.contains(c, na=False)]['listed_in']:
        genre = re.split(r', \s*', i)
        genres.append(genre)
    
    for sublist in genres:
      for item in sublist:
        genres_flat.append(item)
    
    
    return genres_flat

pd.DataFrame(genres_by_country('Japan'))[0].value_counts()
```

```python
pd.DataFrame(genres_by_country('United States'))[0].value_counts()
```

```python
import pandas as pd
import re
df = pd.read_csv("netflix_titles.csv")
```

```python
# 俳優 Global
actors = []
actors_flat = []
for i in df.dropna(subset=['cast'])['cast']:
    actor = re.split(r', \s*', i)
    actors.append(actor)

for sublist in actors:
    for item in sublist:
        actors_flat.append(item)
        
pd.DataFrame(actors_flat)[0].value_counts().head(10)
```

```python
# 俳優 by Country
def actors_by_country(c):
    actors = []
    actors_flat = []
    country_df = df[df.country.str.contains(c, na=False)]
    for i in country_df.dropna(subset=['cast'])['cast']:
        actor = re.split(r', \s*', i)
        actors.append(actor)

    for sublist in actors:
        for item in sublist:
            actors_flat.append(item)
    
    return actors_flat

pd.DataFrame(actors_by_country('Japan'))[0].value_counts().tail(30)
```
